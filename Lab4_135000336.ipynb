```python
import numpy as np
import matplotlib.pyplot as plt


# Define system parameters
frequencies = np.array([0.2, 0.58, 1, 1.7, 2.4]) # input signal frequencies in GHz
fs = 10 # sampling frequency in GHz
N = 100000 # number of samples
sigma2 = 0.1**2 # variance of sampling error


# Generate input signal
t = np.arange(N) / fs

abc = frequencies.reshape(-1, 1)
print("abc", abc)

x = np.sin(2*np.pi*frequencies.reshape(-1, 1)*t).sum(axis=0)


print("x", x)


# Generate ADC output
q = 0.1 # quantization step size
y = np.round(x / q) * q

# Design FIR filter
M_values = np.arange(2, 11) # number of filter taps
print("M values", M_values)
var_ratio = np.zeros_like(M_values)


for i, M in enumerate(M_values):
   # Generate H and E matrices
   H = np.zeros((N-M, M))
   for j in range(M):
    H[:, j] = y[M-j-1:N-j-1]
    E = y[M:N] - x[M:N]

# Compute filter coefficients using least squares
W = np.linalg.pinv(H.T @ H) @ H.T @ E

# Compute estimated sampling error and corrected signal
e_hat = H @ W
y_corr = y[M:] + e_hat


# Compute error signal
E = y_corr - x[M:]

# Compute variance ratio
var_ratio[i] = np.var(E) / (sigma2 / 12)


# Plot variance ratio vs M


plt.plot(M_values, var_ratio)
plt.xlabel('Taps')

plt.ylabel('Variance of Error / Variance of Quantization noise')

plt.title('Ratio of variance of E to Q vs M number of taps')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Given values
fs = 10e9  # Sampling frequency 10 GHz
ts = 1 / fs  # Sampling period
tau = 10e-12  # Time constant 10 ps

# Time vector for one period of the highest frequency 2.4 GHz
t = np.arange(0, 1/2.4e9, ts/100)

# Multi-tone signal frequencies
frequencies = [0.2e9, 0.58e9, 1e9, 1.7e9, 2.4e9]

# Construct the multi-tone signal
multi_tone_signal = np.sum([np.sin(2 * np.pi * f * t) for f in frequencies], axis=0)

# First order model of a ZOH sampling circuit
def sample_and_hold(signal, fs, tau, t):
    sampled_signal = np.zeros_like(signal)
    for n in range(1, len(signal)):
        if n % (1/fs/ts) == 0:
            sampled_signal[n] = sampled_signal[n-1] * np.exp(-ts/tau) + (1 - np.exp(-ts/tau)) * signal[n]
        else:
            sampled_signal[n] = sampled_signal[n-1] * np.exp(-ts/tau)
    return sampled_signal

# Sample the multi-tone signal using the ZOH model
sampled_signal = sample_and_hold(multi_tone_signal, fs, tau, t)

# Plot the original and sampled signals
plt.figure(figsize=(14, 6))
plt.plot(t, multi_tone_signal, label='Original Multi-Tone Signal')
plt.stem(t[::int(1/fs/ts)], sampled_signal[::int(1/fs/ts)], 'r', markerfmt='ro', label='Sampled Signal')
plt.title('Original and Sampled Multi-Tone Signals UIN_135000336')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import lfilter, firwin

# Constants
fs = 10e9  # Sampling frequency (10 GHz)
N_bits = 7  # ADC resolution
V_full_scale = 1.0  # Full scale range (1V)
LSB = V_full_scale / (2 ** N_bits)  # Least Significant Bit
A = 0.5  # NRZ amplitude (0.5V)
T = 1 / fs  # Bit period

# Solve for tau
def calculate_tau():
    tau = -T / np.log(LSB / A)
    return tau

# Compute tau value
tau = calculate_tau()
print(f"Required time constant (tau): {tau:.2e} s")

# Time vector
t = np.arange(0, 100 * T, T)

# Generate NRZ input signal
NRZ_signal = np.random.choice([-A, A], size=len(t))

# Simulate 2-way TI-ADC with mismatches
def ti_adc_simulation(NRZ_signal, time_skew, offset_mismatch, bandwidth_mismatch, noise_std):
    # Two interleaved channels
    ch1_samples = NRZ_signal[::2] + offset_mismatch[0]
    ch2_samples = NRZ_signal[1::2] + offset_mismatch[1]

    # Apply bandwidth mismatch (modeled as first-order LPF)
    b1 = firwin(5, bandwidth_mismatch[0])
    b2 = firwin(5, bandwidth_mismatch[1])
    ch1_samples = lfilter(b1, 1.0, ch1_samples)
    ch2_samples = lfilter(b2, 1.0, ch2_samples)

    # Introduce time skew (sampling delay)
    time_shift = int(time_skew * fs)
    ch2_samples = np.roll(ch2_samples, shift=time_shift)

    # Recombine interleaved samples
    reconstructed_signal = np.zeros(len(NRZ_signal))
    reconstructed_signal[::2] = ch1_samples
    reconstructed_signal[1::2] = ch2_samples

    # Add noise to adjust SNDR
    noise = np.random.normal(0, noise_std, size=len(reconstructed_signal))
    reconstructed_signal += noise

    return reconstructed_signal

# Mismatch parameters
time_skew = 1e-12  # 1 ps skew
offset_mismatch = [0.002, -0.002]  # 2 mV offset mismatch
bandwidth_mismatch = [0.45, 0.55]  # Slight mismatch in cutoff frequencies

# Target SNDR
desired_SNDR_dB = 20
signal_power = np.var(NRZ_signal)

# Iterative noise adjustment
noise_std = 0.001  # Initial guess
tolerance = 0.1  # Tolerance in dB
max_iterations = 20  # Avoid infinite loops

for _ in range(max_iterations):
    adc_output = ti_adc_simulation(NRZ_signal, time_skew, offset_mismatch, bandwidth_mismatch, noise_std)
    error_power = np.var(NRZ_signal - adc_output)
    SNDR = 10 * np.log10(signal_power / error_power)

    if abs(SNDR - desired_SNDR_dB) < tolerance:
        break  # Stop if SNDR is within tolerance

    # Adjust noise level
    noise_std *= np.sqrt((10 ** (desired_SNDR_dB / 10)) / (signal_power / error_power))

print(f"Final SNDR: {SNDR:.2f} dB (Target: {desired_SNDR_dB} dB)")

# Plot results
plt.figure()
plt.plot(t[:200], NRZ_signal[:200], label='Original NRZ Signal')
plt.plot(t[:200], adc_output[:200], label='Reconstructed TI-ADC Output', linestyle='dashed')
plt.legend()
plt.title("NRZ Signal vs. TI-ADC Output")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude (V)")
plt.grid()
plt.show()
```
